```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as grid_spec
import pyplot_themes as themes
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
themes.theme_few(scheme="light", grid=True)
```

# Read Data

```python
data = pd.read_csv('data/spotify_music.csv', index_col = 0)\
    .assign(track_number = lambda x: x['track_number'].astype(int))
```

# Some High Level Stats

```python
print(f"There are {len(data.album.unique())} albums in the data")
```

```python
average_n_of_track_in_albums = int(data.groupby('album')['track_number'].max().reset_index()['track_number'].mean())
print(f"The average number of tracks in albums: {average_n_of_track_in_albums}")
```

```python
data.columns
```

```python
q_cols = ['acousticness','danceability', 'energy', 'instrumentalness', 'liveness', 'loudness','speechiness', 'tempo', 'valence', 'popularity']
for c in q_cols:
    plt.figure(c)
    sns.distplot(data[c])
```

# Correlations

```python
cor_data = data[['acousticness', 
                 'danceability', 'energy',
                 'instrumentalness', 'liveness',
                 'loudness', 'speechiness',
                 'tempo', 'valence', 'popularity']].corr()
```

```python
ax = sns.heatmap(cor_data)
```

There aren't too many strong correlations in the data. The only insights are that:

1. acoustic songs have less "energy" and "dancibility" and are not "loud"
2. Popularity is lightly positively correlated with loudness, danceability and energy
